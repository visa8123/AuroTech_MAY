```python
class ATM:
    def __init__(self,pin,balance):
        self.pin = pin
        self.balance = balance
    def check_balance(self):
        return self.balance
    def withdraw(self,amount):
        if amount > self.balance:
            print("Insufficient balance")
        else:
            self.balance -= amount
            print("Amount has withdrawn successfully!")
    def deposit(self,amount):
        self.balance += amount
        return f"Deposit successful.Thank you for using our Bank"
    def change_pin(self,new_pin):
        self.pin = new_pin
        return f"PIN has changed successfully"
    def transfer_money(self,account_no,amount):
        if amount > self.balance:
            print("Insufficient balance")
        else:
            self.balance -= amount
            return f"{amount} has been successfully transferred to account number {account_no}"

        
    
        
```

```python
def main():
    atm = ATM(6465,5000)

   
    while True :
        print("Welcome to our ATM service.")
        pin = int(input("please enter your PIN:"))
        if pin != atm.pin:
            print("Incorrect PIN> please try again.")
            continue

        print("/n Menu")
        print("1. Check Balance")
        print("2. Withdraw")
        print("3. Deposit")
        print("4. Change PIN")
        print("5. Transfer Money")
        print("6. Exit")
        
        option = int(input("\nChoose one option:"))

        if option == 1:
            print("current balance:",atm.check_balance())
        elif option == 2:
            amount = float(input("Enter the amount to withdraw:"))
            print(atm.withdraw(amount))
        elif option == 3:
            amount = float(input("Enter the amount to deposit:"))
            print(atm.deposit(amount))
        elif option == 4:
            new_pin = int(input("Enter your new PIN:"))
            print(atm.change_pin(new_pin))
        elif option == 5:
            account_no = int(input("Enter account number for transfer:"))
            amount = int(input("Enter the amount to transfer:"))
            print(atm.transfer_money(account_no,amount))
        elif option == 6:
            print("Thank You For Using Our ATM Machine!")
            break
        else:
            print("Invalid option.Please try again.")

```

```python
import tkinter as tk
from tkinter import messagebox

class ATM_GUI:
    def __init__(self, master):
        self.master = master
        self.master.title("ATM Machine")

        self.atm = ATM(6465, 5000)

        self.create_widgets()

    def create_widgets(self):
        self.pin_label = tk.Label(self.master, text="Enter PIN:", font=("Arial", 14))
        self.pin_label.grid(row=0, column=0, pady=10)
        self.pin_entry = tk.Entry(self.master, show="*", font=("Arial", 14))
        self.pin_entry.grid(row=0, column=1, pady=10)

        self.pin_button = tk.Button(self.master, text="Enter", command=self.authenticate_pin, font=("Arial", 12))
        self.pin_button.grid(row=0, column=2, pady=10)

        self.menu_label = tk.Label(self.master, text="Menu:", font=("Arial", 16, "bold"))
        self.menu_label.grid(row=1, column=0, columnspan=3, pady=10)

        self.balance_button = tk.Button(self.master, text="Check Balance", command=self.check_balance, font=("Arial", 12))
        self.balance_button.grid(row=2, column=0, columnspan=3, pady=5)

        self.withdraw_button = tk.Button(self.master, text="Withdraw", command=self.withdraw, font=("Arial", 12))
        self.withdraw_button.grid(row=3, column=0, columnspan=3, pady=5)

        self.deposit_button = tk.Button(self.master, text="Deposit", command=self.deposit, font=("Arial", 12))
        self.deposit_button.grid(row=4, column=0, columnspan=3, pady=5)

        self.change_pin_button = tk.Button(self.master, text="Change PIN", command=self.change_pin, font=("Arial", 12))
        self.change_pin_button.grid(row=5, column=0, columnspan=3, pady=5)

        self.transfer_button = tk.Button(self.master, text="Transfer Money", command=self.transfer_money, font=("Arial", 12))
        self.transfer_button.grid(row=6, column=0, columnspan=3, pady=5)

    def authenticate_pin(self):
        entered_pin = self.pin_entry.get()
        if entered_pin == str(self.atm.pin):
            messagebox.showinfo("Success", "PIN authenticated successfully!")
        else:
            messagebox.showerror("Error", "Incorrect PIN. Please try again.")

    def check_balance(self):
        messagebox.showinfo("Balance", f"Current balance: ${self.atm.check_balance()}")

    def withdraw(self):
        amount = float(input("Enter the amount to withdraw:"))
        self.atm.withdraw(amount)
        messagebox.showinfo("Withdraw", "Amount has been withdrawn successfully!")

    def deposit(self):
        amount = float(input("Enter the amount to deposit:"))
        self.atm.deposit(amount)
        messagebox.showinfo("Deposit", "Deposit successful. Thank you for using our Bank")

    def change_pin(self):
        new_pin = int(input("Enter your new PIN:"))
        self.atm.change_pin(new_pin)
        messagebox.showinfo("PIN Change", "PIN has been changed successfully")

    def transfer_money(self):
        account_no = int(input("Enter account number for transfer:"))
        amount = float(input("Enter the amount to transfer:"))
        message = self.atm.transfer_money(account_no, amount)
        messagebox.showinfo("Transfer Money", message)

class ATM:
    def __init__(self, pin, balance):
        self.pin = pin
        self.balance = balance

    def check_balance(self):
        return self.balance

    def withdraw(self, amount):
        if amount > self.balance:
            messagebox.showerror("Error", "Insufficient balance")
        else:
            self.balance -= amount
            messagebox.showinfo("Success", "Amount has been withdrawn successfully!")

    def deposit(self, amount):
        self.balance += amount
        messagebox.showinfo("Success", "Deposit successful. Thank you for using our Bank")

    def change_pin(self, new_pin):
        self.pin = new_pin
        messagebox.showinfo("Success", "PIN has been changed successfully")

    def transfer_money(self, account_no, amount):
        if amount > self.balance:
            messagebox.showerror("Error", "Insufficient balance")
        else:
            self.balance -= amount
            return f"{amount} has been successfully transferred to account number {account_no}"

if __name__ == "__main__":
    root = tk.Tk()
    atm_gui = ATM_GUI(root)
    root.mainloop()
```
